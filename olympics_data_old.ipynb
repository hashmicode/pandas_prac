```python
import pandas as pd
df = pd.read_csv('olympics.csv')
df.head()
```

```python
df1 = pd.read_csv('olympics.csv',index_col = 0, skiprows=1)
df1.head()
```

```python
df1.columns
```

```python
for column in df1.columns:
    if column[:1] == '01':
        df1.rename(columns={column:'Gold' + column[4:]}, inplace=True)
    if column[:2]=='02':
        df1.rename(columns={column:'Silver' + column[4:]}, inplace=True)
    if column[:2]=='03':
        df1.rename(columns={column:'Bronze' + column[4:]}, inplace=True)
    if column[:1]=='№':
        df1.rename(columns={column:'#' + column[1:]}, inplace=True) 
df1.head()
```
